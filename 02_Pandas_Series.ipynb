#### Working with Series

```python
# Series from a list
list1 = [10, 20, 30, 40]
labels = ['a', 'b', 'c', 'd']
```

```python
series = pd.Series(list1)
type(series)
```

```python
series
```

```python
len(series)
```

```python
print(series[2])
```

```python
series = pd.Series(data=list1)
print(series)
```

```python
series = pd.Series(data=list1, index=labels)
print(series)
```

```python
print(series['c']) # position of 'c' does not matter, it will return the value associated with the label 'c'
```

```python
population = pd.Series(index=['Canada', 'France', 'Japan', 'USA'], data=[40, 65, 125, 340])
print(population)
```

```python
print(population['USA'])
```

```python
# Series from a dictionary, keys become the index and values become the data
dict1 = {'Canada': 40, 'France': 65, 'Japan': 125, 'USA': 340}
series_dict = pd.Series(dict1)
print(series_dict)
```

```python
# Series from a Numpy array
arr = np.array([10, 20, 30, 40])
series_from_array1 = pd.Series(arr)
print(series_from_array1)
```

```python
# Series from a Numpy array and custom index
arr = np.array([10, 20, 30, 40])
series_from_array2 = pd.Series(data=arr, index=labels)
print(series_from_array2)
```

```python
fruits1 = pd.Series({'banana': 8, 'orange': 4, 'apple': 5})
fruits1
```

```python
fruits1.sort_index()
```

```python
fruits1.sort_values() # Ascending order by default
```

```python
fruits1.sort_values(ascending=False) # Descending order
```

```python
fruits1
```

```python
fruits2 = pd.Series({'banana': 5, 'orange': 6, 'apple': 3, 'pineapple': 2})
fruits2
```

```python
fruits = fruits1 + fruits2
fruits
```

```python
# returns a new Series with 3 added to each value in the original Series
fruits.add(3)
```

```python
# returns a new Series with 3 added to each value in the original Series, and fills missing values with 0
fruits.add(3, fill_value=0) 
```

```python
fruits
```

```python
fruits.isnull() # returns a Series of boolean values indicating whether each value is null (NaN) or not
```

```python
fruits.notnull() # returns a Series of boolean values indicating whether each value is not null (NaN) or not
```

```python
# Filtering Series using boolean indexing
fruits[fruits > 8] # returns a Series with only the values greater than 8
```

```python
fruits3 = fruits.notnull()
fruits3
```

```python
fruits
```

```python
fruits3[fruits3 == True].index.to_list() # returns a list of the index labels where the values are True
```

```python
fruits.fillna(2) # returns a new Series with missing values filled with 0
```

```python
fruits
```

```python
fruits.fillna(2, inplace=True) # fills missing values with 2 in the original Series
fruits
```

```python
fruits.to_list() # returns a list of the values in the Series
```

```python
fruits.to_dict() # returns a dictionary with the index labels as keys and the values as values
```

```python
fruits.to_csv('fruits.csv', header=['count'], index_label='fruit')
print('CSV File saved')
```
